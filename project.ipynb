```python
df = df.dropna(subset=['RAM', 'Battery Capacity', 'Screen Size', 'Launched Price (India)'])

# Create a value score (higher = better value)
df['Value Score'] = (df['RAM'] * df['Battery Capacity'] * df['Screen Size']) / df['Launched Price (India)']
```

```python
top_value_phones = df.sort_values(by='Value Score', ascending=False).head(10)

# Show important columns
top_value_phones[['Company Name', 'Model Name', 'RAM', 'Battery Capacity', 'Screen Size', 'Launched Price (India)','Value Score']]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_value_phones, x='Model Name', y='Value Score', hue='Company Name')
plt.title("Top 10 Best Value Mobiles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def clean_price(value):
    value = str(value)
    value = value.replace("INR", "")
    value = value.replace("â‚¹", "")
    value = value.replace(",", "")
    value = value.strip()
    return float(value)

df['Launched Price (India)'] = df['Launched Price (India)'].apply(clean_price)


sns.scatterplot(data=df, x='RAM', y='Launched Price (India)', hue='Company Name')
plt.title("RAM vs Launched Price (India)")
plt.xlabel("RAM (GB)")
plt.ylabel("Price (â‚¹)")
plt.grid(True)
plt.show()
```

```python
df['RAM'] = df['RAM'].astype(str).str.extract(r'(\d+)')     # Extract digits
df['RAM'] = pd.to_numeric(df['RAM'], errors='coerce')        # Convert to float

# Clean the 'Battery Capacity' column
df['Battery Capacity'] = df['Battery Capacity'].astype(str).str.replace(',', '', regex=False)
df['Battery Capacity'] = df['Battery Capacity'].str.extract(r'(\d+)')
df['Battery Capacity'] = pd.to_numeric(df['Battery Capacity'], errors='coerce')

# Clean the 'Screen Size' column
df['Screen Size'] = df['Screen Size'].astype(str).str.extract(r'([\d.]+)')
df['Screen Size'] = pd.to_numeric(df['Screen Size'], errors='coerce')

# Clean the 'Launched Price (India)' column
df['Launched Price (India)'] = df['Launched Price (India)'].astype(str).str.replace('INR', '', regex=False)
df['Launched Price (India)'] = df['Launched Price (India)'].str.replace(',', '', regex=False)
df['Launched Price (India)'] = df['Launched Price (India)'].str.extract(r'(\d+)')
df['Launched Price (India)'] = pd.to_numeric(df['Launched Price (India)'], errors='coerce')
```

```python
df['Battery Capacity'] = df['Battery Capacity'].str.extract(r'(\d+)').astype(int)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Company Name', y='Battery Capacity', data=df, hue='Company Name', palette='pastel')
plt.title("Battery Capacity by Brand")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
